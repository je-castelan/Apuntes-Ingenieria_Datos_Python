```python
import pandas as pd
```

```python
clean_eluniversal = pd.read_csv('Data_Wrangling/eluniversal_2019_04_21_cleaned_articles.csv')
clean_elpais = pd.read_csv('Data_Wrangling/elpais_2019_04_21_cleaned_articles.csv')
```

```python
#Visualización de valores estadísticos de los dataframes limpios
clean_eluniversal.describe() #Hace el estadístico sobre los valores numéricos
```

```python
clean_elpais.describe()
```

```python
#Graficación por puntos
clean_elpais['n_tokens_title'].plot(style='k.') #K= negro. EL punto es que se hace la gráfica de puntos
clean_eluniversal['n_tokens_title'].plot(style='r.') #r= rojo. EL punto es que se hace la gráfica de puntos
```

```python

clean_elpais['n_tokens_body'].plot(style='k.') #K= negro. EL punto es que se hace la gráfica de puntos
clean_eluniversal['n_tokens_body'].plot(style='r.') #r= rojo. EL punto es que se hace la gráfica de puntos
```

```python
#Concatenar dataframes
all_newspapers = pd.concat([clean_eluniversal, clean_elpais])
```

```python
#Agrupador de dataframes por un campo definido
grouped = all_newspapers.groupby('newspaper_uid')
```

```python
#Creación de histogramas
grouped.hist()
```

```python
#Agregaciones de un campo 
grouped['n_tokens_body'].agg(['min','mean','max'])
```

```python
grouped.plot() #Gráfica de líneas con respecto a la agregación
```
